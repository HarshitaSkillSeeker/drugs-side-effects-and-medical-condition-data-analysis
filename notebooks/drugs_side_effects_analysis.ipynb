**Load the Dataset**

```python
# Read a CSV file into a pandas DataFrame from a specified file path
df = pd.read_csv('D:\\UnifiedMentor\\drugs-side-effects-and-medical-condition-data-analysis\\data\\drugs_side_effects_drugs_com.csv')

# Display the first few rows of the DataFrame to verify the data
df.head()
```

**Data Cleaning**

# Check for missing values

```python
# Check for missing values in the DataFrame and return the count of null values for each column
df.isnull().sum()
```

```python
# Fill missing data (categorical data) in the 'side_effects', 'drug_classes','related_drugs','rx_otc','pregnancy_category' column with the string 'Unknown'

df['side_effects'].fillna('Unknown', inplace=True)
df['drug_classes'].fillna('Unknown', inplace=True)
df['related_drugs'].fillna('Unknown', inplace=True)
df['rx_otc'].fillna('Unknown', inplace=True)
df['pregnancy_category'].fillna('Unknown', inplace=True)
```

```python
# Handle alcohol column: NaN -> 0 , 'X' -> 1

df['alcohol'].fillna(0, inplace=True)
df['alcohol'].replace('X', 1, inplace=True)
```

```python
# Fill missing rating and no_of_reviews with 0

df['rating'].fillna(0, inplace=True)
df['no_of_reviews'].fillna(0, inplace=True)
```

```python
# Drop 'brand_names' column

df.drop(columns=['brand_names'], inplace=True)
```

# Common side effects

```python
# Import the Counter class from the collections module (not used in this code snippet)
from collections import Counter

# Import the re module for regular expressions
import re

# Define a function to extract side effects from a string
def extract_side_effects(text):
    # Split the input text into substrings based on the ';' character and strip leading/trailing whitespace
    return [s.strip() for s in re.split(r'[;]', str(text))]

# Apply the extract_side_effects function to each value in the 'side_effects' column and explode the resulting lists into separate rows
side_effect_series = df['side_effects'].apply(extract_side_effects).explode()

# Count the occurrences of each unique side effect and return the top 10 side effects with the highest counts
side_effect_counts = side_effect_series.value_counts().head(10)

# Plot the top 10 side effects in a horizontal bar chart
side_effect_counts.plot(kind='barh')
```

```python
import pandas as pd
import re

# Function to split side effects (semicolon-based split)
def extract_side_effects(text):
    return [effect.strip() for effect in re.split(r'[;]', str(text))]

# Explode dataset
df_exploded = df.copy()
df_exploded['side_effects'] = df_exploded['side_effects'].fillna('Unknown')
df_exploded = df_exploded.assign(
    side_effects_list=df_exploded['side_effects'].apply(extract_side_effects)
)
df_exploded = df_exploded.explode('side_effects_list')

# Group by side_effect × drug and get average rating
heatmap_data = df_exploded.groupby(
    ['side_effects_list', 'drug_name']
)['rating'].mean().reset_index()

# Rename columns for Tableau
heatmap_data.columns = ['side_effect', 'drug_name', 'avg_rating']

# Save to CSV
heatmap_data.to_csv("D:\\UnifiedMentor\\drugs-side-effects-and-medical-condition-data-analysis\\data\\heatmap_rating_side_effect.csv", index=False)
```
